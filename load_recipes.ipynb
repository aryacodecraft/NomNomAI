```python
import ast
import json

# convert string lists → actual lists
df['ingredient_lines'] = df['ingredient_lines'].apply(ast.literal_eval)
df['diet_labels'] = df['diet_labels'].apply(ast.literal_eval)
df['cuisine_type'] = df['cuisine_type'].apply(ast.literal_eval)

# convert JSON → dict
df['total_nutrients'] = df['total_nutrients'].apply(json.loads)
```

```python
def extract_macros(nutrients):
    return {
        "protein": nutrients.get('PROCNT', {}).get('quantity', 0),
        "fat": nutrients.get('FAT', {}).get('quantity', 0),
        "carbs": nutrients.get('CHOCDF', {}).get('quantity', 0)
    }

df[['protein', 'fat', 'carbs']] = df['total_nutrients'].apply(
    lambda x: pd.Series(extract_macros(x))
)
```
